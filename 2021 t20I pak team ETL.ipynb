```python
from selenium import webdriver
```

```python
player_ids = [4425, 3101, 4653, 4412, 4423, 4230, 4747, 6046, 5663, 4991, 4579, 4798, 4536, 5771, 6058]
# player_ids = [5771, 6046]
player_bios = []
```

```python
len(player_ids)
```

```python
for player_id in player_ids:
    driver = webdriver.Chrome("E:\softwares\chromedriver_win32 (1)\chromedriver.exe")
    if player_id not in (6058, 6046) :
        url = "http://www.howstat.com/cricket/Statistics/Players/PlayerOverview_T20.asp?PlayerID=" + str(player_id)
    else:
        url = "http://www.howstat.com/cricket/Statistics/Players/PlayerOverviewSummary.asp?PlayerID=" + str(player_id)
    driver.get(url)
    print(url)
    player_info_fields = driver.find_elements_by_xpath('/html/body/table/tbody/tr[2]/td[3]/table[2]/tbody/tr/td/table[1]//*[@class="FieldName"]')
    table_ref = "/html/body/table/tbody/tr[2]/td[3]/table[2]/tbody/tr/td/table[1]"
    if not player_info_fields:
        print("player_info_fields not found")
        player_info_fields = driver.find_elements_by_xpath('/html/body/div[1]/table/tbody/tr/td/table/tbody/tr/td/table[1]//*[@class="FieldName2"]')
        table_ref = "/html/body/div[1]/table/tbody/tr/td/table/tbody/tr/td/table[1]"
    bio = {}
    for player in player_info_fields:
        field_name = player.text.strip()[:-1]
        print("field_name: ", field_name)
        value = driver.find_element_by_xpath(f'{table_ref}//tr[td[contains(text(),"{player.text.strip()}")]]/td[2]').text
        print("value: ", value)
        bio[field_name] = value
    player_bios.append(bio)
    driver.close()
# bio_values = []
# for field in bio_fields:
#     bio_values.append(driver.find_element_by_xpath(f'/html/body/table/tbody/tr[2]/td[3]/table[2]/tbody/tr/td/table[1]//tr[td[contains(text(),"{field}")]]/td[2]').text)
```

```python
import csv
csv_columns = ['Full Name','Born','Current Age', 'Bats', 'Bowls', 'Matches']
csv_file = "players_bio.csv"
try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in player_bios:
            writer.writerow(data)
except IOError:
    print("I/O error")
```
